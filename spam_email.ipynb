```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train  = df_train.drop('Surname', axis= 1)
df_test  = df_test.drop('Surname', axis= 1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
for i in df_train.columns:
    if df_train[i].dtype == object:
        df_train[i] = label.fit_transform(df_train[i])
        df_test[i] = label.transform(df_test[i])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
X =  df_train.drop('Exited', axis = 1)
y = df_train['Exited']
```

```python
dt = DecisionTreeClassifier()
ab = AdaBoostClassifier()
gb = GradientBoostingClassifier()
rf = RandomForestClassifier()
```

```python
classifiers = [dt, ab, gb, rf]
classifier_names = ['Decision Tree', 'AdaBoost', 'Gradient Boosting', 'Random Forest']
```

```python
for clf, name in zip(classifiers, classifier_names):
    scores = cross_val_score(clf, X, y, cv=5, scoring='accuracy')
    print(f'{name}: Accuracy: {scores.mean()} (± {scores.std()})')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],  
    'max_features': ['auto', 'sqrt', 'log2'],  
    'max_depth': [10, 20, 30],  

}
```

```python
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, verbose=3)
```

```python
grid_search.fit(X,y)
```

```python
y_pred = grid_search.predict(df_test)
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub['Exited'] = y_pred
```
